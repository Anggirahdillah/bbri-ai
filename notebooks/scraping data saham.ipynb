```python
# Import library
import yfinance as yf
import pandas as pd
import os
```

```python
def scrape_stock(stock, start_date, end_date, adjusted=True):
    try:
        print(f"üîç Mengambil data saham {stock} ...")
        df_stock = yf.download(
            stock,
            start=start_date,
            end=end_date,
            auto_adjust=adjusted,
            progress=False
        )

        # Cek kalau data kosong
        if df_stock.empty:
            print(f"‚ùå Gagal: Tidak ada data untuk {stock}. Cek simbol atau tanggal.")
            return

        df_stock = df_stock.reset_index()

        # Handle MultiIndex column
        if isinstance(df_stock.columns, pd.MultiIndex):
            df_stock.columns = df_stock.columns.droplevel(level=1)

        # Lowercase
        df_stock.columns = df_stock.columns.str.lower()

        # Buat folder
        os.makedirs("Saham", exist_ok=True)

        # Nama file
        stock_name = stock.replace(".", "_").lower()
        filename = f"Saham/data_saham_{stock_name}.csv"
        df_stock.to_csv(filename, index=False)

        print(f"‚úÖ Berhasil! Total {len(df_stock)} baris data disimpan ke {filename}")

    except Exception as e:
        print(f"‚ùå Error mengambil data {stock}: {e}")

scrape_stock("BBRI.JK", "2020-01-01", "2025-11-28")
```
